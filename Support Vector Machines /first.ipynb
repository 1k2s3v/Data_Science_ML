```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("mouse_viral_study.csv")
```

```python
df.head()
```

```python
sns.scatterplot(x='Med_1_mL',y='Med_2_mL', hue='Virus Present', data= df)

#Hyperlane (2d Laine)
x=np.linspace(0,10,100)
m=-1
b=11
y=m*x+b

plt.plot(x,y,color='black')
```

```python
from sklearn.svm import SVC
```

```python
help(SVC)
```

```python
y = df['Virus Present']
```

```python
X= df.drop('Virus Present', axis=1 )
```

```python
model = SVC(kernel='linear', C=1000)
```

```python
model.fit(X,y)
```

```python
from svm_margin_plot import plot_svm_boundary
```

график для демонстрации модели 

```python
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel='linear', C=0.005)
```

чем меньше значение С тем больше предел допустиых значений MARGIN 

```python
model.fit(X,y)
```

```python
plot_svm_boundary(model, X,y)
```

```python
model = SVC(kernel='rbf', C= 1, gamma= '')
```

гамма отображает насколько много шума буждет собирать предел допустимых значений  и так же линию разделяющую график 


```python
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model=SVC(kernel='sigmoid')
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
model = SVC(kernel = 'poly', C=0.05, degree=4)
model.fit(X,y)
plot_svm_boundary(model,X,y)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
svm = SVC()
param_grid={'C':[0.01,0.1,1], 
'kernel': ['linear','rbf']}
```

```python
grid = GridSearchCV(svm,param_grid)
```

```python
grid.fit(X,y)
```
