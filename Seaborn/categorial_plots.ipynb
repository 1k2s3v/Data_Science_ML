```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
```

```python
df.head()
```

```python
df['division'].value_counts()
```

```python
plt.figure(figsize=(10,4),dpi=200 )
sns.countplot(data=df,x='level of education')
plt.ylim(0,260)
```

```python
plt.figure(figsize=(10,4),dpi=200 )
sns.barplot(data=df,x='level of education',y='salary', estimator=np.mean,ci='sd',hue='division')
plt.legend(bbox_to_anchor=(1.05,1))
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.jointplot(data=df,x="math score",y='reading score', hue='gender')
```

```python
sns.pairplot(df,hue='gender', corner=True)
```

```python
sns.catplot(data=df,x='gender',y='math score', kind='box',hue ='test preparation course' )
```

```python
sns.pairplot(df)
```

```python
g=sns.PairGrid(df, hue='gender')
g= g.map_lower(sns.scatterplot)
g = g.map_upper(sns.kdeplot)
g = g.map(sns.histplot)

g = g.add_legend()
```
